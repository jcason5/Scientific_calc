```python
import math
```

```python
class calculator():
    def __init__(self, input1, input2):
        self.input1=input1
        self.input2=input2
    
    def addition(num1,num2):
        print("Addition", num1+num2)
        
    def subtraction(num1,num2):
        print("Subtraction", num1-num2)
        
    def multiplication(num1,num2):
        print("Multiplication", num1*num2)
    
    def division(num1,num2):
        print("Division", num1/num2)
        
    def sqr_root(num1):
        print("Square root", math.sqrt(num1))
        
    def power(num1,num2):
        print("Power", math.pow(num1,num2))
        
    def radian(num1):
        print("Degrees to Radians", math.radians(num1))
    
    def factorial(num1):
        print("Factorials", math.factorial(num1))
        
    def perm(num1,num2):
        print("Number of possiblities", math.perm(num1,num2))
        
    def cosine(num1):
        print("Cosine", math.acos(num1))
        
    def sine(num1):
        print("Sine", math.asin(num1))
        
    def ceil(num1):
        print("Rounding", math.ceil(num1))
    
    def degrees(num1):
        print("Radians to Degrees", math.degrees(num1))
    
    def fabs(num1):
        print("Absolute value", math.fabs(num1))
    
    def fmod(num1,num2):
        print("Remainder", math.fmod(num1,num2))
    
    def tan(num1):
        print("Tangent", math.tan(num1))
    
    def gcd(num1,num2):
        print("Greatest Common Devisor", math.gcd(num1,num2))
    
    def floor(num1):
        print("Floor", math.floor(num1))
    
    def isfinite(num1):
        print("Finite", math.isfinite(num1))

while True:
    print("Select number")
    print("1: Addition")
    print("2: Subtraction")
    print("3: Multiplication")
    print("4: Division")
    print("5: Square root")
    print("6: Power")
    print("7: Degrees to Radians")
    print("8: Factorial")
    print("9: Number of possiblities")
    print("10: Cosine")
    print("11: Sine")
    print("12: Rounding")
    print("13: Radians to Degrees")
    print("14: Absolute value")
    print("15: Remainder")
    print("16: Tangent")
    print("17: Greatest common devisor")
    print("18: Floor")
    print("19: Finite")
    print("20: Exit")
    choice=int(input("Enter your choice (1-20):"))
    
    if choice==1:
        num1=int(input("Enter first number:"))
        num2=int(input("Enter second number:"))
        calculator.addition(num1,num2)
    if choice==2:
        num1=int(input("Enter first number:"))
        num2=int(input("Enter second number:"))
        calculator.subtraction(num1,num2)
    if choice==3:
        num1=int(input("Enter first number:"))
        num2=int(input("Enter second number:"))
        calculator.multiplication(num1,num2)
    if choice==4:
        num1=int(input("Enter first number:"))
        num2=int(input("Enter second number:"))
        
        if num2==0:
            print("infinity")
        else:
            calculator.division(num1,num2)
    if choice==5:
        num1=int(input("Enter first number:"))
        calculator.sqr_root(num1)
    if choice==6:
        num1=int(input("Enter first number:"))
        num2=int(input("Enter second number:"))
        calculator.power(num1,num2)
    if choice==7:
        num1=int(input("Enter first number:"))
        calculator.radian(num1)
    if choice==8:
        num1=int(input("Enter first number:"))
        calculator.factorial(num1)
    if choice==9:
        num1=int(input("Enter first number:"))
        num2=int(input("Enter second number:"))
        calculator.perm(num1,num2)
    if choice==10:
        num1=int(input("Enter first number:"))
        calculator.acos(num1)
    if choice==11:
        num1=int(input("Enter first number:"))
        calculator.asin(num1)
    if choice==12:
        num1=int(input("Enter first number:"))
        calculator.ceil(num1)
    if choice==13:
        num1=int(input("Enter first number:"))
        calculator.degrees(num1)
    if choice==14:
        num1=int(input("Enter first number:"))
        calculator.fabs(num1)
    if choice==15:
        num1=int(input("Enter first number:"))
        num2=int(input("Enter second number:"))
        calculator.fmod(num1,num2)
    if choice==16:
        num1=int(input("Enter first number:"))
        calculator.tan(num1)
    if choice==17:
        num1=int(input("Enter first number:"))
        num2=int(input("Enter second number:"))
        calculator.gcd(num1,num2)
    if choice==18:
        num1=int(input("Enter first number:"))
        calculator.floor(num1)
    if choice==19:
        num1=int(input("Enter first number:"))
        calculator.isfinite(num1)
    elif choice==20:
        break
    else:
        print("Invalid input. Pick a number 1-20.")
```
